#### 1.1. Importing Libraries

```python
# Let's import the pandas, numpy libraries as pd, and np respectively. 
import pandas as pd
import numpy as np

# Load the pyplot collection of functions from matplotlib, as plt 
import matplotlib.pyplot as plt
```

#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
## Traspose the dataframe, so that we have one row per borough
properties = properties.transpose()

## Reset the index to create a column for Long borough
properties = properties.reset_index()

## Assign values of the first row to column headings, and inspect the dataframe again
new_names = properties.iloc[0]
properties = properties[1:]
properties.columns = new_names
properties.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
## Rename the first two columns
properties = properties.rename(columns = {properties.columns[0]: "London Borough", properties.columns[1]: "ID"})
properties.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
## Melt the dataframe, using 'London Borough' and 'ID' as id variables
## Each row should have the average price for one borough, in one month
properties = properties.melt(id_vars = ['London Borough', 'ID'], var_name = 'Month', value_name = 'Average Price')
properties.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
## Convert Average Price column to floating point numbers
conversion = {'Average Price': 'float'}
properties = properties.astype(conversion)

print(properties.dtypes)
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
## Check for missing values by column
properties.isna().sum()
```

```python
## In the next few lines of code,we show that the columns null values all have 
##'London Borough' equal to 'Unnamed' followed by a number.
## There is no clear way to fill in these missing values.
has_null = properties[properties.isna().any(axis = 1)]
has_null['London Borough'].unique()
```

```python
## Given what we discovered in the previous cell, we delete all rows with missing values.
properties = properties.dropna()

## Next, we examine the contents of the 'London Borough' column
## We notice that many entries do not appear to be valid Borough names
all_names = properties['London Borough'].unique()
print(all_names)
```

```python
## Checking against a list of London Boroughs, we find that all correct borough names found here:
## https://en.wikipedia.org/wiki/London_boroughs.
## We see the valid borough names are listed in order above, starting with 'Barking & Dagenham'
## We subset our data again, keeping only the rows with a valid entry in 'London Borough'.
boroughs = all_names[1:33]
properties = properties[properties['London Borough'].isin(boroughs)]
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
## Before plotting the data, we extra 'Year' from the date column,
## and calculate the average price for each year.
properties['Year'] = properties['Month'].apply(lambda x: x.year)
prop_by_year = properties.groupby(['London Borough', 'ID', 'Year']).mean()
prop_by_year = prop_by_year.reset_index()

### We restrict to data from the date range 1998-2018, and examine our data again.
prop_by_year = prop_by_year.loc[(prop_by_year['Year'] >= 1998) & (prop_by_year['Year'] <= 2018)]
prop_by_year.head()
```

```python
## Average price data for a single London Borough shows a large increase over the period in question
bex_df = prop_by_year.loc[prop_by_year['London Borough'] == 'Bexley']
timerange = bex_df['Year'].unique()
bex_df.plot(x = 'Year', y = 'Average Price', rot = '45', title = 'Average Housing Prices, Bexley')
plt.xticks(timerange, rotation = 45)
plt.show()
```

```python
## To get a better sense of the data, we divide the boroughs into four quartiles, based on average housing price in 1998.
## We will plot average price by year on the same axes for all boroughs in each quartile.
starter = prop_by_year.loc[prop_by_year['Year'] == 1998]
starter = starter.sort_values('Average Price')
sorted_boroughs = starter['London Borough']
```

```python
plt.rcParams["figure.figsize"] = (12, 12)

timerange = prop_by_year['Year'].unique()

## Define a function 
def plot_borough(borough, time_range):
    my_prices = prop_by_year.loc[prop_by_year['London Borough'] == borough]['Average Price']
    plt.plot(timerange, my_prices, label = borough)

for borough in sorted_boroughs[:8]:
    plot_borough(borough, timerange)

ax = plt.gca()
plt.title('Bottom Quartile by Average 1998 Price')
plt.xlabel("Year")
plt.xticks(timerange, rotation = 45)
plt.ylabel("Average Price")
plt.legend()
plt.show()
```

```python
for borough in sorted_boroughs[8:16]:
    plot_borough(borough, timerange)
   
ax = plt.gca()
plt.title('Second Quartile by Average 1998 Price')
plt.xlabel("Year")
plt.xticks(timerange, rotation = 45)
plt.ylabel("Average Price")
plt.legend()
plt.show()
```

```python
for borough in sorted_boroughs[16:24]:
    plot_borough(borough, timerange)

ax = plt.gca()
plt.title('Third Quartile by Average 1998 Price')
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.xticks(timerange, rotation = 45)
plt.legend()
plt.show()
```

```python
for borough in sorted_boroughs[24:]:
    plot_borough(borough, timerange)

ax = plt.gca()
plt.title('Top Quartile by Average 1998 Price')
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.xticks(timerange, rotation = 45)
plt.legend()
plt.show()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
### A function which calculates price ratio for a given borough.
def get_price_ratio(borough):
    condition_dates = (prop_by_year['London Borough'] == borough) & (prop_by_year['Year'].isin([1998, 2018]))
    small_df = prop_by_year[condition_dates].sort_values('Year')
    return (small_df.iloc[1,3]/small_df.iloc[0,3])

ratio_list = [(borough, get_price_ratio(borough)) for borough in boroughs]
ratio_data = pd.DataFrame(ratio_list, columns = ('London Borough', 'Price Ratio'))
ratio_data = ratio_data.sort_values('Price Ratio', ascending = False)

print("Table of Top 5 Boroughs by Price Ratio:")
ratio_data.head()
```
